```python
# 문제 2-3. 주문 메서드

# 음료의 온도에 따라 temp 변수를 "HOT" 또는 "ICE"로 지정하세요.

class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    def menu_select(self):
        # [여기 추가되었어요] 주문 내역과 가격을 저장할 빈 리스트 준비
        self.order_menu = []
        self.order_price = []

        n = -1

        # [여기 변경되었어요] 0을 입력할 때까지 주문을 반복하는 루프
        while n != 0:
            print()
            n = int(input("음료 번호를 입력하세요 (주문 완료는 0) : "))

            if 1 <= n <= len(self.menu):
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                # [여기 추가되었어요] 선택한 메뉴와 가격을 리스트 끝에 추가
                self.order_menu.append(self.temp + ' ' + self.menu[n - 1])
                self.order_price.append(self.price[n - 1])

                print(f"주문 음료: {self.temp} {self.menu[n - 1]} : {self.price[n - 1]}원 추가되었습니다.")
                print('현재 장바구니:', self.order_menu)

            elif n != 0:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # [여기 추가되었어요] 반복이 끝난 후(0 입력 시) 최종 내역 출력
        print("\n최종 주문 내역:", self.order_menu)
        print("최종 가격 내역:", self.order_price)
```

```python
# 아이스 초코라떼 주문해보기
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()
```

```python
# 문제 3-3. 클래스 업데이트
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요!

class Kiosk:
    def __init__(self):
        # [[YOUR CODE]]
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    # 메뉴 출력 메서드
    def menu_print(self):
        # [[YOUR CODE]]
        for i in range(len(self.menu)):
            print(f"{i + 1}. {self.menu[i]} : {self.price[i]}원")

    # 주문 메서드
    def menu_select(self):
        # [[YOUR CODE]]
        self.order_menu = []
        self.order_price = []
        n = -1

        while n != 0:
            print()
            n = int(input("음료 번호를 입력하세요 (주문 완료는 0) : "))

            if 1 <= n <= len(self.menu):
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT(1) 또는 ICE(2)를 선택하세요 : "))
                    if t == 1: self.temp = "HOT"
                    elif t == 2: self.temp = "ICE"
                    else: print("1 또는 2를 입력해주세요.")

                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.order_price.append(self.price[n-1])
                print(f"-> {self.temp} {self.menu[n-1]} 추가되었습니다.")

            elif n != 0:
                print("잘못된 번호입니다. 다시 입력해주세요.")

    # 지불
    def pay(self):
        # [[YOUR CODE]]
        payment = input("\n지불 수단을 선택하세요 (1.현금 / 2.카드) : ")
        if payment in ['1', 'cash']:
            print('직원을 호출하겠습니다.')
        elif payment in ['2', 'card']:
            print('IC칩 방향에 맞게 카드를 꽂아주세요.')
        else:
            print('결제 방식을 다시 확인해주세요.')

    # 주문서 출력
    def table(self):
        # [[YOUR CODE]]
        print('\n' + '╓' + '-' * 30 + '╖')
        for i in range(3): print('|' + ' ' * 31 + '|')

        for i in range(len(self.order_menu)):
            print(f" {self.order_menu[i]}  :  {self.order_price[i]}")

        print(f"\n [합계 금액 : {sum(self.order_price)}원]")
        for i in range(3): print('|' + ' ' * 31 + '|')
        print('╙' + '-' * 30 + '╜')
```

```python
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
